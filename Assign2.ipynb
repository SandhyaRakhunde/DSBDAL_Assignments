```python
#handling outlires by Z-Score Method

upper = df['CGPA'].mean()+df['CGPA'].std()

lower = df['CGPA'].mean()-df['CGPA'].std()
final = df[(df['CGPA']<upper) & (df['CGPA']>lower)]
```

```python
# using Interquartile range methode to remove the outliers
n = int(df.shape[0]/4);
m= int(3*df.shape[0]/4);
result = df.iloc[:,5:]
result1 = result.sort_values("CGPA",ignore_index=True)
q1 = result.at[n-1,"CGPA"]
q3 = result.at[m-1,"CGPA"]
iqr = q3-q1
low = q1-1.5*iqr
high = q3+1.5*iqr
df1 = df.copy()  #copying the df dataframe into df1

df.loc[df['CGPA']>high,'CGPA'] = high
df.loc[df['CGPA']<low,'CGPA'] = low
df
```
