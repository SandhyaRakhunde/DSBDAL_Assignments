```python
#for numeric datatype replacing null with mean()
df[['Distance', 'Postcode','Bedroom2','Bathroom','Car','BuildingArea','YearBuilt','Lattitude','Longtitude','Propertycount']] = df[['Distance', 'Postcode','Bedroom2','Bathroom','Car','BuildingArea','YearBuilt','Lattitude','Longtitude','Propertycount']].fillna(df[['Distance', 'Postcode','Bedroom2','Bathroom','Car','BuildingArea','YearBuilt','Lattitude','Longtitude','Propertycount']].mean())
```

```python
# Handling null values using df.fillna()

#for integer => replacing null values from 3 columns by that column mean
df['Landsize'].fillna(int(df['Landsize'].mean()), inplace=True)   
# df['BuildingArea'].fillna(int(df['BuildingArea'].mean()), inplace=True)
# df['YearBuilt'].fillna(int(df['YearBuilt'].mean()), inplace=True)

#for string datatype replacing NaN with another string
df['CouncilArea'].fillna(' No CouncilArea', inplace=True)
df['Regionname'].fillna(' No Regionname', inplace=True)
```

```python
# min-max Normalization
# as huge difference between min and max values => to bring them to normal scale => using normalization

# Landsize 
df = df.assign(Landsize=(df.Landsize - df['Landsize'].min())/(df['Landsize'].max() - df['Landsize'].min()))

# BuildingArea
df = df.assign(BuildingArea=(df.BuildingArea - df['BuildingArea'].min())/(df['BuildingArea'].max() - df['BuildingArea'].min()))

# Propertycount
df = df.assign(Propertycount=(df.Propertycount - df['Propertycount'].min())/(df['Propertycount'].max() - df['Propertycount'].min()))
```
